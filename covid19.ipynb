```python
import matplotlib.pyplot as plt
x=df.iloc[61:,0]
y=df.iloc[61:,1]
plt.figure(figsize=(20,8))
ax=plt.axes()
ax.grid(color='#f8f8f8',linewidth=1.0)
ax.set_facecolor('black')
ax.set_xlabel('Date',color='red',size=40)
ax.set_ylabel('No.of Cases',color='green',size=40)
plt.xticks(rotation='vertical',color='red',size=10)
plt.tick_params(color='white',size=20)
for i,j in zip(x,y):
    ax.annotate(j,xy=(i,j),color='white',size=12)
ax.annotate('Second Lockdown April 15 2021',
            xy=(15.2,860),
            xytext=(19.2,500),
            color='white',
            size=25,
            arrowprops=dict(color='white',linewidth=1.0))
plt.plot(x,y,marker='o',ms=12)
plt.title('Covid19 Second lockdown',size=50)
plt.savefig('covid19.pdf')
plt.show
```
